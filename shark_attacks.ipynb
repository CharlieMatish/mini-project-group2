```python
def filter_and_replace(df, Activity, SURFING):
    
    df[Activity] = df[Activity].apply(lambda x: SURFING if SURFING in x else x)
    return df

df = filter_and_replace(df, 'Activity', 'SURFING')
df
```

```python
def filter_and_replace(df, Activity, SWIMMING):
    
    df[Activity] = df[Activity].apply(lambda x: SWIMMING if SWIMMING in x else x)
    return df

df = filter_and_replace(df, 'Activity', 'SWIMMING')
df
```

```python
def filter_and_replace(df, Activity, DIVING):
    
    df[Activity] = df[Activity].apply(lambda x: DIVING if DIVING in x else x)
    return df

df = filter_and_replace(df, 'Activity', 'DIVING')
df
```

```python
def filter_and_replace(df, Sex, M):
    
    df[Sex] = df[Sex].apply(lambda x: M if M in x else x)
    return df

df = filter_and_replace(df, 'Sex', 'M')
df
```
